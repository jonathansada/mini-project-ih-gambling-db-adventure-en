```python
import pandas as pd
import sqlite3 
```

```python
# read xlxs

# Tabs to import: Betting, Product, Customer, and Account
betting_df = pd.read_excel("dataset/SQL Test Data.xlsx", sheet_name="Betting", usecols="A:I")
ac_bet_df = pd.read_excel("dataset/SQL Test Data.xlsx", sheet_name="Betting", usecols="M:O")
ac_prod_df = pd.read_excel("dataset/SQL Test Data.xlsx", sheet_name="Betting", usecols="Q:W,Y")
product_df = pd.read_excel("dataset/SQL Test Data.xlsx", sheet_name="Product")
customer_df = pd.read_excel("dataset/SQL Test Data.xlsx", sheet_name="Customer")
accounts_df = pd.read_excel("dataset/SQL Test Data.xlsx", sheet_name="Account")

ac_bet_df.dropna(how="all", axis=0, inplace=True)
ac_prod_df.dropna(how="all", axis=0, inplace=True)

display(betting_df)
display(ac_bet_df)
display(ac_prod_df)
display(product_df)
display(customer_df)
display(accounts_df)
```

```python
# EXPORT TO DB
account_dtype = {'AccountNo': 'TEXT', 'CustId':'INTEGER', 'AccountLocation':'TEXT', 'CurrencyCode':'TEXT','DailyDepositLimit':'INTEGER', 'StakeScale':'INTEGER', 'SourceProd':'TEXT'}
customer_dtype = {'CustId':'INTEGER', 'AccountLocation':'TEXT', 'Title':'TEXT', 'FirstName':'TEXT', 'LastName':'TEXT','CreateDate':'INTEGER', 'CountryCode':'TEXT', 'Language':'TEXT', 'Status':'TEXT', 'DateOfBirth':'INTEGER', 'Contact':'TEXT', 'CustomerGroup':'TEXT'}
betting_dtype = {'AccountNo':'INTEGER', 'BetDate':'INTEGER', 'ClassId':'TEXT', 'CategoryId':'INTEGER', 'Source':'TEXT', 'BetCount':'INTEGER', 'Bet_Amt':'REAL', 'Win_Amt':'REAL', 'Product':'TEXT'}
ac_bet_dtype = {'AccountNo.1':'TEXT', 'Bet_Amt.1':'REAL', 'Product.1':'TEXT'}
ac_prod_dtype = {'AccountNo.2':'TEXT', 'Vegas':'REAL', 'Sportsbook':'REAL', 'Games':'REAL', 'Casino':'REAL', 'Poker':'REAL', 'Bingo':'REAL', 'Adjustments':'REAL'}
product_dtype = {'CLASSID':'TEXT', 'CATEGORYID':'INTEGER', 'product':'TEXT', 'sub_product':'TEXT', 'description':'TEXT', 'bet_or_play':'BOOL'}


con = sqlite3.connect('output/gambling.db')
accounts_df.to_sql(name="Account", con=con, if_exists="replace", index=False, index_label="AccountNo", dtype=account_dtype)
customer_df.to_sql(name="Customer", con=con, if_exists="replace", index=False, index_label="CustId", dtype=customer_dtype)
betting_df.to_sql(name="Betting", con=con, if_exists="replace", index=False, dtype=betting_dtype)
ac_bet_df.to_sql(name="AccBet", con=con, if_exists="replace", index=False, dtype=ac_bet_dtype)
ac_prod_df.to_sql(name="AccProd", con=con, if_exists="replace", index=False, dtype=ac_prod_dtype)
product_df.to_sql(name="Product", con=con, if_exists="replace", index=False, dtype=product_dtype)
```
